```python
import requests
import pandas as pd
import time

API_KEY = "f8e3284a3d0d0b081393486e295ba704"
titles = []
years = []
ratings = []
release_dates = []
vote_counts = []
popularity = []
original_language = []
overview = []

unique_movies = set()

for page_num in range(1, 30):  # Pages 1-23
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={API_KEY}&page={page_num}&language=en-US"
    
    try:
        response = requests.get(url)
        
        if response.status_code != 200:
            print(f"  Error: Status code {response.status_code}")
            break
        
        data = response.json()
        movies = data.get('results', [])
        
        if not movies:
           
            break
     
        for movie in movies:
            title = movie.get('title', 'N/A')
            release_date = movie.get('release_date', 'N/A')
            year = release_date.split('-')[0] if release_date and release_date != 'N/A' else 'N/A'
            rating = movie.get('vote_average', 'N/A')
            vote_count = movie.get('vote_count', 'N/A')
            popularity_score = movie.get('popularity', 'N/A')
            language = movie.get('original_language', 'N/A')
            movie_overview = movie.get('overview', 'N/A')
            
            unique_id = (title, year)
            if unique_id not in unique_movies:
                titles.append(title)
                years.append(year)
                ratings.append(rating)
                release_dates.append(release_date)
                vote_counts.append(vote_count)
                popularity.append(popularity_score)
                original_language.append(language)
                overview.append(movie_overview)
                
                unique_movies.add(unique_id)
 
        if len(titles) >= 600:
            break
        
        time.sleep(0.3) 
    
    except Exception as e:
        print(f"  Error fetching page {page_num}: {e}")
        continue

df = pd.DataFrame({
    "title": titles,
    "year": years,
    "rating": ratings,
    "vote_count": vote_counts,
    "release_date": release_dates,
    "popularity": popularity,
    "language": original_language,
    "overview": overview
})
df
```

```python
df.to_csv('top_rated_movies.csv', index=False, encoding='utf-8')
```
