```python
initial_shape = df.shape[0]

# Removig movies with invalid years (e.g., future years or too old)
current_year = 2025
df = df[(df['year'] >= 1900) & (df['year'] <= current_year)]

# Removing movies with rating outside valid range (0-10)
df = df[(df['rating'] >= 0) & (df['rating'] <= 10)]

# Removing movies with negative vote counts
df = df[df['vote_count'] > 0]

final_shape = df.shape[0]
removed = initial_shape - final_shape

print(f"Rows before cleaning: {initial_shape}")
print(f"Rows after cleaning: {final_shape}")
print(f"Rows removed: {removed}")
```

```python
#Pivot: Decade vs Language (Average Rating)
```

```python
if 'decade' not in df.columns:
    df['decade'] = (df['year'] // 10) * 10
#print("PIVOT TABLE: DECADE vs LANGUAGE (Average Rating - Top 5 Languages):")
top_5_languages = df['language'].value_counts().head(5).index
df_top_lang = df[df['language'].isin(top_5_languages)]

pivot_decade_lang = pd.pivot_table(df_top_lang, 
                                   values='rating', 
                                   index='decade', 
                                   columns='language', 
                                   aggfunc='mean', 
                                   fill_value=0).round(2)
#print(pivot_decade_lang)
# Visualization - Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_decade_lang, annot=True, fmt='.2f', cmap='YlOrRd', 
            linewidths=0.5, cbar_kws={'label': 'Average Rating'})
plt.xlabel('Language', fontsize=12, fontweight='bold')
plt.ylabel('Decade', fontsize=12, fontweight='bold')
plt.title('Average Rating by Decade and Language (Heatmap)', fontsize=16, fontweight='bold')
plt.yticks([i + 0.5 for i in range(len(pivot_decade_lang))], 
           [f"{int(d)}s" for d in pivot_decade_lang.index], rotation=0)
plt.tight_layout()
plt.show()
```

```python
#rating_category column will be created 
if 'rating_category' not in df.columns:
    def categorize_rating(rating):
        if rating >= 8.5:
            return 'Excellent'
        elif rating >= 8.0:
            return 'Very Good'
        elif rating >= 7.5:
            return 'Good'
        else:
            return 'Average'
    
    df['rating_category'] = df['rating'].apply(categorize_rating)
    #print("Rating Category column created!")

#popularity_category column created
if 'popularity_category' not in df.columns:
    df['popularity_category'] = pd.cut(df['popularity'], 
                                        bins=[0, 5, 10, 20, 100], 
                                        labels=['Low', 'Medium', 'High', 'Very High'])
    #print("Popularity Category column created!")

#print("CROSSTAB: RATING CATEGORY vs POPULARITY CATEGORY (Percentages):")
crosstab_rating_pop_pct = pd.crosstab(df['rating_category'], 
                                      df['popularity_category'], 
                                      normalize='index') * 100
crosstab_rating_pop_pct = crosstab_rating_pop_pct.round(2)
#print(crosstab_rating_pop_pct)

# Visualization - Stacked Bar (Percentage)
plt.figure(figsize=(12, 7))
crosstab_rating_pop_pct.plot(kind='bar', stacked=True,
                             color=['lightgreen', 'yellow', 'orange', 'red'],
                             edgecolor='black', alpha=0.8)
plt.xlabel('Rating Category', fontsize=12, fontweight='bold')
plt.ylabel('Percentage (%)', fontsize=12, fontweight='bold')
plt.title('Popularity Category Distribution within Each Rating Category (%)', 
          fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Popularity Category', fontsize=10, title_fontsize=11)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
#SCATTER PLOTS - Multiple Variable Relationships
# Year vs Popularity (colored by Rating)
plt.figure(figsize=(14, 7))
scatter = plt.scatter(df['year'], df['popularity'], 
                     c=df['rating'], cmap='RdYlGn', 
                     s=100, alpha=0.6, edgecolors='black', linewidth=0.5)
plt.colorbar(scatter, label='Rating')
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Popularity', fontsize=12, fontweight='bold')
plt.title('Year vs Popularity (Color = Rating)', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
