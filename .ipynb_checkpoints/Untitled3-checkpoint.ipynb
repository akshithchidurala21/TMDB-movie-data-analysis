```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

```python
url='https://www.imdb.com/chart/top/?ref_=hm_nv_menu'
```

```python
request_header = {'Content-Type': 'application/json;charset=UTF-8', 
                  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36',
                'Accept-Encoding': 'gzip, deflate, br, zstd',
                  'Accept-language':'en-US,en;q=0.9,hi;q=0.8,zh-CN;q=0.7,zh;q=0.6',
                  'referer':'https://www.imdb.com/chart/top/?ref_=hm_nv_menu'
                 }

response = requests.get(url, headers=request_header)
response
```

```python
html_text=response.text
```

```python
soup=BeautifulSoup(html_text)
soup.text
```

```python

titles = []
years = []
ratings = []
genres = []
runtimes = []
directors = []
stars = []
votes = []
```

```python


    soup = BeautifulSoup(response.text, 'html.parser')
    movies = soup.select('.sc-fc35a1ef-0 hTMtRz')
    
    for movie in movies:
        # Title
        title = movie.select_all('did', class_= 'ipc-title__text ipc-title__text--reduced')
        titles.append(title.text.strip() if title else '')
        # Year
        year = movie.select_one('span.lister-item-year')
        years.append(year.text.strip() if year else '')
        # IMDb Rating
        rating = movie.select_one('div.inline-block.ratings-imdb-rating strong')
        ratings.append(rating.text.strip() if rating else '')
        # Genre
        genre = movie.select_one('span.genre')
        genres.append(genre.text.strip() if genre else '')
        # Runtime
        runtime = movie.select_one('span.runtime')
        runtimes.append(runtime.text.strip() if runtime else '')
        # Votes
        vote = movie.select_one('span[name="nv"]')
        votes.append(vote.text.strip() if vote else '')
        # Directors and Stars (optional, shown in p text under .text-muted)
        p_text = movie.select_one('p.text-muted ~ p')
        if p_text:
            people = p_text.text.split('|')
            directors.append(people[0].replace("Director:", "").strip() if "Director:" in people[0] else '')
            stars.append(people[1].replace("Stars:", "").strip() if len(people) > 1 else '')
        else:
            directors.append('')
            stars.append('')

    time.sleep(2)  # Courtesy delay

# Build DataFrame
df = pd.DataFrame({
    'Title': titles,
    'Year': years,
    'IMDb_Rating': ratings,
    'Genre': genres,
    'Runtime': runtimes,
    'Votes': votes,
    'Director': directors,
    'Stars': stars
})

print(df.head())
```
