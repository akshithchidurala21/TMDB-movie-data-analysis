```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

url = "https://ev-database.org/#group=vehicle-group&rs-pr=10000_100000&rs-er=0_1000&rs-ld=0_1000&rs-ac=2_23&rs-dcfc=0_400&rs-ub=10_200&rs-tw=0_2500&rs-ef=100_350&rs-sa=-1_5&rs-w=1000_3500&rs-c=0_5000&rs-y=2010_2030&s=1&p=0-10"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
print("Status code:", response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')

# Placeholder lists for each car attribute
car_titles = []
availability = []
ranges = []
efficiency = []
weights = []
acceleration = []
battery = []
fastcharge = []
towing = []
cargo = []
prices = []

# Regex to help extract fields
for section in soup.find_all("div"):
    text = section.get_text(separator=" ", strip=True)
    # Simple logic: look for anchor fields
    if "Range" in text and "Battery" in text and "€" in text:
        # Title: get up to newline or pattern
        title_match = re.search(r'^([A-Za-z0-9][^\n]*)', text)
        car_titles.append(title_match.group(1).strip() if title_match else '')

        # Availability
        avail = re.search(r'Available to order since .*?\d{4}', text)
        if not avail:
            avail = re.search(r'Discontinued.*?\d{4}', text)
        availability.append(avail.group(0) if avail else '')

        # Extract numeric fields
        rng = re.search(r'Range\*?\s*(\d+)\s*km', text)
        ranges.append(rng.group(1) if rng else '')

        eff = re.search(r'Efficiency\*?\s*(\d+)\s*Wh/km', text)
        efficiency.append(eff.group(1) if eff else '')

        wt = re.search(r'Weight\s*(\d+)\s*kg', text)
        weights.append(wt.group(1) if wt else '')

        acc = re.search(r'0-100\s*(\d+\.?\d*)\s*sec', text)
        acceleration.append(acc.group(1) if acc else '')

        batt = re.search(r'Battery\*?\s*(\d+\.?\d*)\s*kWh', text)
        battery.append(batt.group(1) if batt else '')

        fast = re.search(r'Fastcharge\s*(\d+)\s*kW', text)
        fastcharge.append(fast.group(1) if fast else '')

        tow = re.search(r'Towing\s*(\d+)\s*kg', text)
        towing.append(tow.group(1) if tow else '')

        cargo_vol = re.search(r'Cargo Vol.\s*(\d+)\s*L', text)
        cargo.append(cargo_vol.group(1) if cargo_vol else '')

        price = re.search(r'€(\d[\d,\.]+)', text)
        prices.append(price.group(1) if price else '')

# Compose row tuples for full deduplication
rows = list(zip(car_titles, availability, ranges, efficiency, weights, acceleration, battery, fastcharge, towing, cargo, prices))
unique_rows = list(set(rows))  # Remove row duplicates

# Convert back to DataFrame
df = pd.DataFrame(unique_rows, columns=[
    "car_title", "availability", "range_km", "efficiency_Wh_per_km", "weight_kg",
    "acceleration_0_100_sec", "battery_kWh", "fastcharge_kW", "towing_kg", "cargo_L", "price_eur"
])

# Reset index for clean DataFrame
df = df.reset_index(drop=True)
df
```
